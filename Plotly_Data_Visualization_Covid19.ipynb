```python
import plotly.graph_objects as px
# interactive visualization
import plotly.express as pex
```

```python
def world_map_plot_func(df, col, color):
    df = df[df[col]>0]
    fig = pex.choropleth(df, locations="Country/Region", locationmode='country names', 
                  color=col, hover_name="Country/Region", 
                  title=col, hover_data=[col], color_continuous_scale=color)
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    fig.show()
```

```python
world_map_plot_func(df, 'Confirmed', 'speed')
```

```python
 
plot = px.Figure(data=[px.Scatter(
    x=df['Country/Region'],
    y=df['Confirmed'],
    mode='markers',)
])
 
# Add dropdown
plot.update_layout(
    updatemenus=[
        dict(buttons=list([
            dict(
                args=["type", "scatter"],
                label="Scatter Plot",
                method="restyle"
            ),
            dict(
                args=["type", "bar"],
                label="Bar Chart",
                method="restyle"
            ),
             dict(
                args=["type", "box"], 
                label="Box Plot",
                method="restyle"
            ),
             dict(
                args=["type", "pie"],
                label="Pie Plot",
                method="restyle"
            )
        ]),
            direction="down",
            
        ),
    ]
    
)
 
plot.show()
```

```python

def button_func(counties,col):
  plot = px.Figure(data=[px.Scatter(
      x=counties ,
      y=col,
      mode='markers',
      )
  ])
  
  # Add dropdown
  plot.update_layout(
      updatemenus=[
          dict(
              type="buttons",
              direction="left",
              buttons=list([
                  dict(
                      args=["type", "scatter"],
                      label="Scatter Plot",
                      method="restyle"
                  ),
                  dict(
                      args=["type", "box"],
                      label="Box Plot",
                      method="restyle"
                  )
              ]),
          ),
      ]
  )
  
  plot.show()
```

```python
button_func(df['Country/Region'],df['Confirmed'])
```

```python
button_func(df['Country/Region'],df['Active'])
```

```python
fig = pex.scatter(df.sort_values('Deaths', ascending=False).iloc[:20, :], 
                 x='Confirmed', y='Deaths', color='Country/Region', size='Confirmed', 
                 height=700, text='Country/Region', log_x=True, log_y=True, 
                 title='Deaths vs Confirmed')
fig.update_traces(textposition='top center')
fig.update_layout(showlegend=False)
fig.update_layout(xaxis_rangeslider_visible=True)
fig.show()
```

```python
def bar_plot_horz(df, col, n, hover_data=[]):
    fig = pex.bar(df.sort_values(col).tail(n), 
                 x=col, y="Country/Region", color='WHO Region',  
                 text=col, orientation='h', width=700, hover_data=hover_data,
                 color_discrete_sequence = pex.colors.sequential.Blackbody) #pex.colors.cyclical.mygbm) #pex.colors.qualitative.Dark2)
    fig.update_layout(title=col, xaxis_title="", yaxis_title="", 
                      yaxis_categoryorder = 'total ascending',
                      uniformtext_minsize=8, uniformtext_mode='hide')
    fig.show()
```

```python
for i in range(len(list1)):
  bar_plot_horz(df, list1[i], 15)
```

```python
def tree_map_plot_func(col):
    fig = pex.treemap(df, path=["Country/Region"], values=col, height=700,
                 title=col, color_discrete_sequence =  pex.colors.cyclical.mygbm)
    fig.data[0].textinfo = 'label+text+value'
    fig.show()
```

```python
for i in range(len(list1)):
  tree_map_plot_func(list1[i])
```

```python
# color pallette
cnf, dth, rec, act = '#393e46', '#ff2e63', '#21bf73', '#fe9801' 
```

```python
df_selected_colums = df_day.groupby('Date')['Confirmed','Recovered', 'Deaths', 'Active'].sum().reset_index()
df_selected_colums = df_selected_colums.melt(id_vars="Date", value_vars=['Confirmed','Recovered', 'Deaths', 'Active'],
                 var_name='Case', value_name='Count')
df_selected_colums.head()

fig = pex.area(df_selected_colums, x="Date", y="Count", color='Case', height=600, width=700,
             title='Cases over time', color_discrete_sequence = [rec, dth, act,cnf])
fig.update_layout(xaxis_rangeslider_visible=True)
fig.show()
```

```python
df_country=pd.read_csv('nouman_covid_19_Dataset.csv')
```

```python
def bar_plot_stacked_values(col):
    fig = pex.bar(df_country, x="Date", y=col, color='Country/Region', 
                 height=600, title=col, 
                 color_discrete_sequence =pex.colors.qualitative.Dark2)
    fig.update_layout(showlegend=True)
    fig.show()
```

```python
for i in range(len(list1)):
  bar_plot_stacked_values(list1[i])
```

```python
for i in range(len(list1)):

  top = df_country[df_country['Date'] == df_country['Date'].max()]
  top_casualities = top.groupby(by = 'Country/Region')[list1[i]].sum().sort_values(ascending = False).head(20).reset_index()
  top_casualities

  plt.figure(figsize= (15,10))
  plt.xticks(fontsize = 15)
  plt.yticks(fontsize = 15)
  plt.xlabel("Total cases",fontsize = 30)
  plt.ylabel('Country',fontsize = 30)
  plt.title(f"Top 20 countries having most {list1[i]} cases" , fontsize = 30)
  ax = sns.barplot(x = top_casualities[list1[i]], y = top_casualities['Country/Region'])
  for i, (value, name) in enumerate(zip(top_casualities[list1[i]],top_casualities['Country/Region'])):
      ax.text(value, i-.05, f'{value:,.0f}',  size=10, ha='left',  va='center')
  ax.set(xlabel='Total cases', ylabel='Country')
```
